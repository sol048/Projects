## Background and Prior Work


Sleep deprivation is a pervasive issue among students, particularly in academic settings where high workloads, extracurricular activities, and social pressures often compete for time. Moreover, poor sleep quality is significantly associated with stress level. Despite its importance, studies indicate that over 60% of college students report insufficient sleep, with many averaging fewer than 7 hours per night.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) This raises significant concerns about how sleep deprivation impacts academic outcomes, especially GPA.

Prior research has established strong correlations between sleep disorders and stress. The studies show that various stress factors significantly cause insomnia<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). More recently, a survey by the American Academy of Sleep Medicine<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) linked poor sleep hygiene to a 30% increase in dropout rates among university students. These studies collectively underscore sleep as a modifiable factor influencing academic achievement. However, gaps remain in understanding how specific aspects of sleep deprivation—such as duration—interact with academic performance, particularly GPA, and stress level. This project builds on prior work by examining not only broad correlations but also granular relationships between sleep metrics and stress.

## Exploratory Data Analysis


## Section 1. Analyzing Sleep Duration and Quality vs. GPA

This section of the code aims to visualize the relationship between **sleep duration, GPA, sleep quality, and stress levels** using the two datasets we introduced earlier. The goal is to explore whether sleep habits correlate with academic success and how stress affects sleep quality.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_a = df
data_b = df2

sns.set_theme(style="whitegrid", palette="muted")
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# Dataset A Sleep vs GPA
sns.regplot(
    data=data_a,
    x="Sleep_Hours_Per_Day",
    y="GPA",
    ax=ax[0],
    scatter_kws={"s": 100, "alpha": 0.7},
    line_kws={"color": "red"}
)
ax[0].set_title("Dataset A: Sleep Hours vs. GPA\n(Students)")
ax[0].set_xlabel("Sleep Hours Per Day (hours)")  # Added units
ax[0].set_ylabel("GPA (0-4 scale)")  # Added scale
ax[0].annotate(
    "Optimal GPA near 7–8 hours",
    xy=(7.5, 3.0),
    xytext=(5, 3.2),
    arrowprops=dict(arrowstyle="->", color="black")
)

# Regression Plot: This plot shows the relationship between sleep duration and GPA using a linear regression model (sns.regplot()).
# Scatter Points: Each point represents a student's sleep hours and corresponding GPA.
# Trend Line: A red regression line helps identify the overall trend.
# Annotation: Highlights an optimal sleep range of 7–8 hours, where students tend to have higher GPAs.


# Dataset B compares sleep Duration vs quality (ONLY Students)
students_b = data_b[data_b["Occupation"] == "Student"]

# Add trend line using regplot first
sns.regplot(
    data=students_b,
    x="Sleep Duration (hours)",
    y="Quality of Sleep (scale: 1-10)",
    scatter=False,
    ax=ax[1],
    line_kws={"color": "red", "linestyle": "--"}
)

# Then plot points with hue
sns.scatterplot(
    data=students_b,
    x="Sleep Duration (hours)",
    y="Quality of Sleep (scale: 1-10)",
    hue="Stress Level (scale: 1-10)",
    ax=ax[1],
    s=100,
    palette="viridis"
)

# Filtering for Students: Since Dataset B includes people from various occupations, we filter for students only before visualization.
# Scatter Plot: This plot compares Sleep Duration vs. Sleep Quality:
# X-axis: Sleep duration (hours)
# Y-axis: Self-reported sleep quality (scale: 1-10)
# Hue (Color Coding): Stress level (scale: 1-10), where darker colors represent higher stress levels.
# Reference Line: A red dashed vertical line marks 7 hours of sleep, typically recommended for optimal sleep.
# Annotation: Notes that high stress levels correlate with lower sleep quality.

ax[1].set_title("Dataset B: Sleep Duration vs. Sleep Quality\n(Students)")
ax[1].axvline(7, color="red", linestyle="--", label="Optimal Sleep")
ax[1].annotate(
    "High stress correlates with poor sleep quality",
    xy=(4.5, 4),
    xytext=(3, 2),
    arrowprops=dict(arrowstyle="->", color="black")
)

# Add explicit legend with scale description
handles, labels = ax[1].get_legend_handles_labels()
ax[1].legend(
    handles, 
    labels,
    title="Stress Level (1-10 scale)",
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.tight_layout()
plt.show()
```

This visualization provides key insights into how sleep duration impacts both GPA and sleep quality:

Optimal sleep duration (~7-8 hours) is associated with higher GPAs.
Students with high stress levels report lower sleep quality, regardless of sleep duration.
Comparing both datasets provides a more generalizable understanding of student sleep patterns and their impact on academic performance.


## Section 2. Analyzing Stress Levels and Sleep Patterns

This section explores how **stress levels affect sleep duration** and the **relationship between stress and sleep disorders**. We use **boxplots and countplots** to visualize trends in both datasets.

```python
## **1. Visualizing Stress Levels vs. Sleep Duration (Dataset A)**
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=data_a,
    x="Stress_Level",
    y="Sleep_Hours_Per_Day",
    order=["Low", "Moderate", "High"]
)
plt.title("Dataset A: Stress Level vs. Sleep Hours (Students)")
plt.xlabel("Stress Level (Categories: Low/Moderate/High)")  # Added category clarification
plt.ylabel("Sleep Hours Per Day (hours)")  # Added units
plt.axhline(7, color="red", linestyle="--", label="Optimal Sleep (7h)")
plt.annotate(
    "High stress linked to shorter sleep",
    xy=(2, 6.5),
    xytext=(1.5, 5),
    arrowprops=dict(arrowstyle="->", color="black")
)
plt.legend()
plt.show()

#Boxplot Analysis: Displays the distribution of sleep hours across different stress levels (Low, Moderate, High). 
#Helps identify median sleep duration, spread, and potential outliers.

#**Key Observations**
#Higher stress levels correlate with shorter sleep duration.
#The red horizontal line at 7 hours marks the optimal sleep duration.
#Highlights that high-stress students tend to sleep less.

plt.figure(figsize=(12, 6))

# Create stress categories with explicit bin labels
data_b["Stress Category"] = pd.cut(
    data_b["Stress Level (scale: 1-10)"],
    bins=[0, 3, 6, 10],
    labels=["Low (1-3)", "Moderate (4-6)", "High (7-10)"]  # Added scale ranges
)
sns.countplot(
    data=data_b,
    x="Stress Category",
    hue="Sleep Disorder",
    order=["Low (1-3)", "Moderate (4-6)", "High (7-10)"]
)
plt.title("Dataset B: Stress Level vs. Sleep Disorders")
plt.xlabel("Stress Level (1-10 scale categories)")  # Added scale reference
plt.ylabel("Count of Students")  # Explicit y-axis label
plt.annotate(
    "High stress linked to insomnia",
    xy=(2, 25),
    xytext=(1, 30),
    arrowprops=dict(arrowstyle="->", color="black")
)

# Add explicit legend title
plt.legend(title="Sleep Disorder Presence")  # Added clarification

plt.tight_layout()
plt.show()

#Countplot Analysis: Groups stress levels into categories (Low, Moderate, High) for consistency across datasets.
# Shows the number of students with/without sleep disorders at each stress level.

#**Key Observations**
#Higher stress levels are associated with a higher prevalence of sleep disorders.
#Annotation highlights a trend that students experiencing high stress are more likely to suffer from sleep-related issues like insomnia.
```

These visualizations provide deeper insights into the impact of stress on sleep habits:

- Students with higher stress levels sleep fewer hours, which may impact cognitive function and academic performance.
- Higher stress correlates with a higher likelihood of sleep disorders, reinforcing the negative effects of stress on sleep quality.
- Optimal sleep (~7 hours) is observed in students with lower stress levels, supporting the need for better stress management.

## Section 3. Analyzing Stress Level and GPA

This section shows visualization of examining the **relationship between the stress level and GPA**. We use boxplot and scatter plot to see the distributions of GPA based on stress level and GPA distribution per stress leve.

```python
# GPA Distribution Analysis: Visualizing the relationship between stress levels and GPA.

# Boxplot: 
# - Displays the distribution of GPA for each stress level category (Low, Moderate, High).
# - Helps identify the median GPA, interquartile range (spread of data), and presence of outliers.
# - Provides a clear view of how GPA varies across stress levels.

#Distribution of GPA based on Stress Level
plt.figure(figsize=(8, 5))
sns.boxplot(x="Stress_Level", y="GPA", data=data_a, order=["Low", "Moderate", "High"])
plt.xlabel("Stress Level")
plt.ylabel("GPA")
plt.title("GPA Distribution by Stress Level")
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()

# Scatter Plot (Stripplot): 
# - Visualizes individual GPA data points within each stress level category.
# - Introduces jitter (random small shifts) to avoid overlapping points, making patterns easier to see.
# - Helps highlight how GPAs are spread within each stress category.

#GPA distribution per Stress Level
plt.figure(figsize=(8, 5))
sns.stripplot(x="Stress_Level", y="GPA", data=data_a, order=["Low", "Moderate", "High"], jitter=True, size=8)
plt.xlabel("Stress Level")
plt.ylabel("GPA")
plt.title("GPA vs Stress Level (Scatter Plot)")
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

These visualizations indicate a positive association between stress level and GPA:

Both visualizations confirm that students with higher GPAs tend to have higher stress levels.

- The boxplot shows a clear upward trend in GPA as stress levels increase, with students in the 'High' stress category having the highest median GPA.
- The scatter plot reinforces this trend by demonstrating a higher concentration of high-GPA students in the 'High' stress category.
- However, there is some variation within each stress level, suggesting that while GPA and stress are related, other factors might also influence this relationship.

### Key Findings

#### 1. Sleep Duration and GPA
The scatter plot and regression analysis showed that while students who sleep between 7–8 hours per night tend to have higher GPAs (~3.2–3.5 range), the overall trend is weakly positive.
- The regression line is nearly flat, indicating a low correlation between sleep and GPA.
- Students who sleep less than 6 hours per night tend to have GPAs clustering around 2.5–3.0, while those who sleep 8+ hours often achieve GPAs above 3.3.
- Despite these trends, the correlation coefficient (r-value) is close to 0, suggesting that sleep alone does not strongly predict GPA.


#### 2. Stress Levels and Sleep Patterns
Boxplots revealed a clear inverse relationship between stress levels and sleep duration:
- Students with low stress (1–3 on a scale of 10) sleep a median of ~8.5 hours per night, well above the recommended 7-hour threshold.
- Those with moderate stress (4–6) sleep around 7 hours, indicating a noticeable decline.
- Students with high stress (7–10) sleep around 6–6.5 hours, with some reporting as low as 5 hours per night.

Furthermore, stress does not just affect sleep duration—it also impacts sleep quality:
- The regression plot of sleep quality vs. sleep duration shows that students who sleep fewer than 6 hours report sleep quality scores below 5 (on a 10-point scale).
- Highly stressed students reported the most sleep disorders, with over 8 cases of insomnia among those in the highest stress category compared to only 3 cases in the moderate group and 1 in the low-stress group.
- This highlights that stress not only shortens sleep but also reduces its restorative quality, contributing to higher fatigue and potential academic burnout.

#### 3. Stress Levels and GPA
Our boxplot and scatter plot of stress vs. GPA showed an unexpected trend:
- Students with higher stress (7–10) actually have higher median GPAs (~3.3) compared to those with low stress (~2.8).
- The GPA distribution in moderate stress levels (4–6) is centered around 3.0–3.2, forming a middle ground.
- While stress negatively affects sleep, this trend suggests that some students might perform better academically under pressure.

However, outliers in the high-stress group show extreme variations—some students achieved near-perfect 4.0 GPAs, while others dropped below 2.5, indicating that excessive stress may either motivate or overwhelm students, depending on individual coping mechanisms.

### Discussion
Stress, sleep, and GPA are closely connected, with stress acting as a key factor that influences both sleep habits and academic performance. Our analysis shows that students with higher stress levels tend to sleep less and report poorer sleep quality, which can negatively impact focus, memory, and overall learning. Despite this, high-stress students often have higher median GPAs, suggesting that stress can sometimes act as a motivator for academic performance. However, the data also reveals a wide range of outcomes among highly stressed students—while some thrive under pressure, others struggle, leading to lower GPAs. ### Limitations and Future Considerations
While our findings provide strong insights, we acknowledge there are several limitations:
- **Self-reported data** may introduce bias, as students may **overestimate or underestimate** their sleep duration and stress levels.
- **Correlation does not imply causation**—other factors such as **study habits, major difficulty, and lifestyle choices** could also impact GPA.
- **GPA alone is an incomplete metric** of academic success, as it does not reflect comprehension, creativity, or long-term knowledge retention.

Future research should use objective sleep-tracking devices (e.g., smartwatches, EEG monitors) and longitudinal studies to observe changes over time. Additionally, experimental studies on stress reduction techniques (e.g., mindfulness programs, structured sleep interventions) could help determine causal effects.

Our findings emphasize the need for schools to address sleep and stress management to enhance student well-being and academic success. Some potential interventions include:
- Providing mental health and stress-reduction programs,such as therapy and guided relaxation sessions.
- Encouraging better sleep habit by educating students on the dangers of sleep deprivation.

By acknowledging the impact of stress and sleep on performance, educational institutions can create a healthier, more supportive learning environment that fosters both academic excellence and student well-being.
